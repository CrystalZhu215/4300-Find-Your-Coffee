```python
import csv
import pandas as pd
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from collections import Counter
import string
```

```python
df = pd.read_csv('coffee_fix.csv')
df
```

```python
stemmed_words = []
stem_to_original = {}

stemmer = PorterStemmer()

combined_descriptions = df[["desc_1", "desc_2", "desc_3"]].apply(lambda x : ' '.join(x.dropna()), axis=1)

for text in combined_descriptions:
    tokens = word_tokenize(text)
    for token in tokens:
        if all(char.isalnum() or char in string.whitespace for char in token):
            token = token.lower()
            stemmed_word = stemmer.stem(token)
            stemmed_words.append(stemmed_word)
            if len(token) > 1:
                if stemmed_word not in stem_to_original:
                    stem_to_original[stemmed_word] = set(token)
                else:
                    stem_to_original[stemmed_word].add(token)
```

```python
print(stemmed_words[:50])
```

```python
word_counts = Counter(stemmed_words)

sorted_word_counts = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)

print("Top 10 most common stemmed words:")
for word, count in sorted_word_counts[:100]:
    print(f"!!{word}: {count}")
    if len(word) > 1:
        for original_word in stem_to_original[word]:
            if len(original_word) > 1:
                print(original_word)
    print()
```
