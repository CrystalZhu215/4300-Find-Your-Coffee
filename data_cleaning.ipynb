Standardizing the prices

```python
#split the est_price text 
df['est_price_split'] = df['est_price'].str.split()
```

```python
# HKD is in the text, append it into a list
hkd_indices = []
for i in range(len(df)):
    row = df.iloc[i]
    if 'HKD' in row['est_price_split']:
        hkd_indices.append(i)
hkd_data= df.iloc[hkd_indices]
```

```python
hkd_data.head()
```

```python
# extracting the first number 
def extract_first_number(row):
    cost = row['est_price_split'][1]

    number_str = cost[cost.index('$') + 1: cost.index('/')]
    return int(number_str)

hkd_data['first_number'] = hkd_data.apply(extract_first_number, axis = 1)
```

```python
#convert HKD/g into USD$ / 12ounces 
hkd_data['first_number_calculated'] = hkd_data['first_number'] * 44.225
```

```python
#extracting the second number
def extract_second_number(row):
    cost = row['est_price_split'][1]
    length_first_number = len(str(row['first_number']))
    number_str = cost[cost.index('$') + length_first_number +2: ]
    return int(number_str)

hkd_data['second_number'] = hkd_data.apply(extract_second_number, axis = 1)
```

```python
# final conversion
hkd_data['calculated_price'] = (hkd_data['first_number_calculated'] / hkd_data['second_number']).round(2)
```

```python
# adding the calculated number into the format of $ / 12ounce
hkd_data['changed_price'] = hkd_data['calculated_price']
for index, row in hkd_data['calculated_price'].iteritems():
    hkd_data['changed_price'][index] = '$' + str(hkd_data['calculated_price'][index]) + '/12 ounces'
```

```python
hkd_data.head()
```

```python
# input the changed price into the actual df
for index in hkd_indices:
    for index, row in hkd_data['changed_price'].iteritems():
        df['est_price'][index] = hkd_data['changed_price'][index]
    
```
