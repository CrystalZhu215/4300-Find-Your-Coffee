```python
df.at[1845, 'est_price'] = 'NT $500/227 grams'
df.at[1914, 'est_price'] = 'NT $350/8 ounces'
df.at[1979, 'est_price'] = 'NT $400/227 grams'
```

```python
#sum of nulls in each column 
df.isnull().sum()
```

```python
# fill null values in roast to Medium-Light
df['roast'] = df['roast'].fillna('Medium-Light')
```

```python
#fill null values to mean of the columns
column_means = df.mean().round()
df = df.fillna(column_means)
```

```python
#fill null values for location to Comayagua, Honduras
df['location'] = df['location'].fillna('Comayagua, Honduras')
```

```python
#rows where the estimated price is null
null_state = df['est_price'].isnull()
for index, row in null_state.iteritems():
    if row == True:
        print(index)
```

```python
# delete these rows because there's no estimated price online (seems like they no longer sell it anymore)
rows_to_delete = [458, 751, 863, 1210, 1753]
df = df.drop(rows_to_delete)
```

```python
# null values for desc_3
null_state = df['desc_3'].isnull()
for index, row in null_state.iteritems():
    if row == True:
        print(index)
```

```python
#set the null values for desc_3 as same description as desc_1
df['desc_3'][207] = df['desc_1'][207]
df['desc_3'][492] = df['desc_1'][492]
```

```python
df.isnull().sum()
```

Standardizing the prices

```python
df['est_price'] = df['est_price'].replace(',', '', regex=True)
df['est_price'] = df['est_price'].replace('-', ' ', regex=True)
```

```python
#split the est_price text 
df['est_price_split'] = df['est_price'].str.split()
```

```python
# grams to ounces
grams_indices = []
for i in range(len(df)):
    row = df.iloc[i]
    if len(row['est_price_split']) > 1 and "grams" in row['est_price_split'][1]:
        grams_indices.append(i)
grams_data= df.iloc[grams_indices]
```

```python
grams_data
```

```python
def extract_first_number_grams(row):
    cost = row['est_price_split'][0]
    if '$' in cost:
        number_str = cost[cost.index('$') + 1: cost.index('/')].strip()
        return number_str
    else:
        number_str = cost[cost.index('¥') + 1: cost.index('/')].strip()
        return number_str


grams_data['first_number'] = grams_data.apply(extract_first_number_grams, axis = 1)
```

```python
#extracting the second number
def extract_second_number_grams(row):
    cost = row['est_price_split'][0]
    length_first_number = len(str(row['first_number']))
    number_str = cost[cost.index('/') +1: ].strip()
    return int(number_str)

grams_data['second_number'] = grams_data.apply(extract_second_number_grams, axis = 1)

```

```python
#final conversion
grams_data['calculated_price'] = grams_data['first_number']
for index, row in grams_data['est_price_split'].iteritems():
    if '$' in grams_data['est_price_split'][index][0]:
        grams_data['calculated_price'][index] = round((float(grams_data['first_number'][index]) * 340.194)/ (float(grams_data['second_number'][index])),2)
    else:
        grams_data['calculated_price'][index] = round((((float(grams_data['first_number'][index]) * 340.194 * 0.0066) / (float(grams_data['second_number'][index])))),2)
```

```python
# adding the calculated number into the format of $ / 12ounce
grams_data['changed_price'] = grams_data['calculated_price']
for index, row in grams_data['calculated_price'].iteritems():
    grams_data['changed_price'][index] = '$' + str(grams_data['calculated_price'][index]) + '/12 ounces'
```

```python
# input the changed price into the actual df
for index, row in grams_data['changed_price'].iteritems():
    df['est_price'][index] = grams_data['changed_price'][index]
```

```python
# HKD is in the text, append it into a list
hkd_indices = []
for i in range(len(df)):
    row = df.iloc[i]
    if 'HKD' in row['est_price_split']:
        hkd_indices.append(i)
hkd_data= df.iloc[hkd_indices]
```

```python
hkd_data.head()
```

```python
# extracting the first number 
def extract_first_number(row):
    cost = row['est_price_split'][1]

    number_str = cost[cost.index('$') + 1: cost.index('/')].strip()
    return number_str

hkd_data['first_number'] = hkd_data.apply(extract_first_number, axis = 1)
```

```python
#convert HKD/g into USD$ / 12ounces 
hkd_data['first_number_calculated'] = hkd_data['first_number'].astype('int32') * 44.225
```

```python
#extracting the second number
def extract_second_number(row):
    cost = row['est_price_split'][1]
    length_first_number = len(str(row['first_number']))
    number_str = cost[cost.index('$') + length_first_number +2: ].strip()
    return int(number_str)

hkd_data['second_number'] = hkd_data.apply(extract_second_number, axis = 1)
```

```python
# final conversion
hkd_data['calculated_price'] = (hkd_data['first_number_calculated'] / hkd_data['second_number']).round(2)
```

```python
# adding the calculated number into the format of $ / 12ounce
hkd_data['changed_price'] = hkd_data['calculated_price']
for index, row in hkd_data['calculated_price'].iteritems():
    hkd_data['changed_price'][index] = '$' + str(hkd_data['calculated_price'][index]) + '/12 ounces'
```
